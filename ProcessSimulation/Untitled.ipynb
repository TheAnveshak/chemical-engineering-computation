```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
F, x_f = 1000, 0.5
D, x_D = 450, None
B, x_B = None, 0.05
```

```python
if D is not None and x_D is not None:
    B   = F - D
    x_B = (F * x_f - D * x_D) / B
    print(f"x_B = {x_B:.6f}")

elif B is not None and x_B is not None:
    D   = F - B
    x_D = (F * x_f - B * x_B) / D
    print(f"x_D = {x_D:.6f}")

elif D is not None and x_B is not None:
    B   = F - D
    x_D = (F * x_f - B * x_B) / D
    print(f"x_D = {x_D:.6f}")

elif B is not None and x_D is not None:
    D   = F - B
    x_B = (F * x_f - D * x_D) / B
    print(f"x_B = {x_B:.6f}")

elif x_D is not None and x_B is not None:
    A = np.array([[1.0,   1.0],
                  [x_D,  x_B]])
    b = np.array([F, F * x_f])
    D, B = np.linalg.solve(A, b)
    print(f"D = {D:.6f},  B = {B:.6f}")

else:
    raise ValueError("Must provide exactly one flow (D or B) and one composition (x_D or x_B),\n"
                     "or both compositions x_D & x_B.")

```

```python
c_p = 4.18 #kJ/kgK
lam = 2250 #kJ/kg
R = 2.5
```

```python
Reflux_flow = R*D
Reflux_flow
```

```python
Boilup_rate = F+Reflux_flow-B
Boilup_rate # kg/hr
```

```python
steam_rate = Boilup_rate * lam/lam
steam_rate
```

```python
energy = steam_rate*lam #kJ/hr
energy = energy / 3600 # kW
energy
```

```python
energy_TR = energy * 3.5168
energy_TR
```

```python
condenser_heat_duty = (R+1)*D * lam
condenser_heat_duty
```

```python
cooling_water_flow = condenser_heat_duty/(4.18 *5)
cooling_water_flow
```

```python
coolin_water_mass = cooling_water_flow/1000
cost = coolin_water_mass * 18
```
